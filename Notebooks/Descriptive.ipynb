```python
datos2008.corr()
```

```python
matriz = datos2008.corr()
plt.figure(figsize=(10,5))
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1, fignum=1, aspect='auto')
plt.xticks(range(len(datos2008.columns)), datos2008.columns, rotation=90)
plt.yticks(range(len(datos2008.columns)), datos2008.columns)

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i,j, round(matriz.iloc[i, j],2), ha="center", va="center")

plt.colorbar()
plt.show()
```
