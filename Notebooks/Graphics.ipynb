```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
dominios = ["Dominio1", "Dominio2", "Dominio3", "Dominio4", "Dominio5"]
valores = [
    [5, 4, 3, 3.5, 4.5],  # Valores en el tiempo 1
    [4, 4.5, 2.5, 4, 4],  # Valores en el tiempo 2
    [4.5, 4.2, 2.7, 3.8, 4.3]  # Valores en el tiempo 3
    # ... (y así sucesivamente para t4, t5, ... t9)
]

N = len(dominios)
theta = np.linspace(0, 2 * np.pi, N, endpoint=False)
theta = np.concatenate((theta, [theta[0]]))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Por cada serie de tiempo, crea un polígono en el gráfico 3D
for t, val in enumerate(valores):
    val += val[:1]  # cierra el polígono
    xs = val * np.cos(theta)
    ys = val * np.sin(theta)
    zs = [t] * (N+1)
    ax.plot(xs, ys, zs, label=f"Tiempo {t+1}")
    ax.fill(xs, ys, zs, alpha=0.2)

# Ajustes
ax.set_title("Gráfica de radar 3D a lo largo del tiempo")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Tiempo")
ax.legend()

plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Categorias y datos (simulados para este ejemplo)
categories = ['Locomoción', 'Vitalidad', 'Psicológico', 'Sensorial', 'Cognitivo']
data = {
    '2021': [0.8, 0.6, 0.7, 0.9, 0.5],
    '2022': [0.7, 0.65, 0.75, 0.85, 0.55],
    '2023': [0.68, 0.67, 0.77, 0.88, 0.6]
}

# Preparar datos para el gráfico 3D
z = np.array([data['2021'], data['2022'], data['2023']])
x, y = np.meshgrid(range(len(categories)), range(3))

# Crear figura
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Ajustar títulos de eje y layout
fig.update_layout(scene=dict(
        xaxis_title='Categoría',
        yaxis_title='Año',
        zaxis_title='Valor',
        xaxis=dict(tickvals=list(range(len(categories))), ticktext=categories),
        yaxis=dict(tickvals=list(range(3)), ticktext=['2021', '2022', '2023'])
    ))

fig.show()
```

```python
import plotly.graph_objects as go

# Datos
categories = ['Locomotion', 'Cognition', 'Vitality', 'Sensory', 'Psychological']
values = [11, 10, 10, 12, 10]

# Colores
colors = ['blue', 'orange', 'green', 'purple', 'red']

# Crear figura
fig = go.Figure()

# Añadir traza para cada categoría
fig.add_trace(go.Scatterpolar(
    r = values,
    theta = categories,
    line=dict(color='peru'),
    fill='toself',
    fillcolor='rgba(0,0,0,0)'  # Transparente para que solo se vea la línea
))

# Actualizar el layout con colores
fig.update_traces(line=dict(width=3))
for i, color in enumerate(colors):
    fig.add_shape(
        go.layout.Shape(
            type="line",
            x0=0.5,
            x1=1,
            y0=0.5,
            y1=0.5,
            xref="paper",
            yref="paper",
            line=dict(
                color=color,
                width=3
            )
        ),
        row="polar",
        col=i+1
    )

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 15]  # Ajustar según el rango deseado
        ),
        angularaxis=dict(
            rotation=90,  # Iniciar en la parte superior
            direction="counterclockwise"
        )
    ),
    showlegend=False  # Ocultar leyenda para una visualización más clara
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
labels = ['Locomotion', 'Cognition', 'Vitality', 'Sensory', 'Psychological']
values = [11, 10, 10, 12, 10]
colors = ['blue', 'orange', 'green', 'purple', 'red']

# Crear un conjunto de ejes polares
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'})

# Calcular ángulos para cada dimensión
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values += values[:1]  # cerrar el polígono
angles += angles[:1]  # cerrar el polígono

# Dibujar datos
ax.fill(angles, values, color='grey', alpha=0.25)
ax.plot(angles, values, color='grey', linestyle='solid', linewidth=0.6)

# Dibujar líneas de colores para cada dimensión
for angle, color in zip(angles, colors):
    ax.plot([angle, angle], [0, 15], color=color, linestyle='solid', linewidth=1.5)

# Añadir etiquetas
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_rlabel_position(115)
ax.set_rticks([2, 4, 6, 8, 10, 12, 14])  # Less radial ticks
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
dominios = ["Locomotion", "Sensory", "Vitality", "Cognition", "Psichological"]
valores_t1 = [12, 12, 12, 12, 12]  # Valores en el tiempo 1
valores_t2 = [10, 9, 9, 11, 8]  # Valores en el tiempo 2
valores_t3 = [9, 9, 8, 12, 12]
valores_t4 = [9, 8, 7, 12, 12]

# Crear un gráfico de radar
N = len(dominios)
theta = np.linspace(0, 2 * np.pi, N, endpoint=False)

# Asegurarse de que la gráfica se cierra volviendo al inicio
valores_t1 += valores_t1[:1]
valores_t2 += valores_t2[:1]
valores_t3 += valores_t3[:1]
valores_t4 += valores_t4[:1]

theta = np.concatenate((theta, [theta[0]]))

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Dibujar datos
ax.plot(theta, valores_t1, label="2008")
ax.plot(theta, valores_t2, label="2012")
ax.plot(theta, valores_t3, label="2016")


# Añadir etiquetas, título y leyenda
ax.set_xticks(theta)
ax.set_xticklabels(dominios + [dominios[0]])  # Añade el primer dominio al final para cerrar el gráfico
ax.set_title("Gráfica de radar de cinco dominios a lo largo del tiempo")
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores
categorias = ['Locomotion', 'Cognition', 'Vitality', 'Sensory', 'Psychological']
valores = [12, 12, 12, 12, 12]

# Número de categorías
num_categorias = len(categorias)

# Ángulos para cada categoría
angulos = np.linspace(0, 2 * np.pi, num_categorias, endpoint=False).tolist()

# Asegurarse de que la figura sea circular
valores += valores[:1]
angulos += angulos[:1]

# Crear gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angulos, valores, color='lightblue', alpha=0.5)

# Configuración de las etiquetas
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(categorias)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Valores
categorias = ['Locomocion', 'Cognition', 'Vitality', 'Sensory', 'Psychological']
valores = [12, 12, 12, 12, 12]

# Número de categorías
num_categorias = len(categorias)

# Ángulos para cada categoría
angulos = np.linspace(0, 2 * np.pi, num_categorias, endpoint=False).tolist()

# Asegurarse de que la figura sea circular
valores += valores[:1]
angulos += angulos[:1]

# Crear figura 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Configuración de las coordenadas x y y
x = np.cos(angulos)
y = np.sin(angulos)

# Configuración de las coordenadas z (valores)
z = np.array(valores)

# Crear el gráfico de radar en 3D
ax.plot(x, y, z, color='skyblue', linewidth=2, linestyle='solid', marker='o', markersize=8, label='Radar Chart')

# Configuración de las etiquetas
ax.set_xticks(x)
ax.set_yticks(y)
ax.set_xticklabels(categorias)

# Mostrar el gráfico
plt.show()
```
